```python
import json
import matplotlib.pyplot as plt
y_true = []
y_predit = []
x = []
cnt = 1
with open("tot_losses.json",'r') as f:
    data = json.load(f)
for i in range(len(data[0][0])):
    y_true.append(data[0][0][i])
    y_predit.append(data[0][1][i])
    x.append(cnt)
    cnt += 1
print(len(data))

plt.plot(x,y_true,color='blue', label='true_loss')

plt.plot(x,y_predit,color='red', label='predict_loss')
```

```python
import json
import matplotlib.pyplot as plt
y_true = []
y_predit = []
x = []
cnt = 1
with open("tot_losses.json",'r') as f:
    data = json.load(f)
for i in range(len(data[0][0])):
    y_true.append(data[9][0][i])
    y_predit.append(data[9][1][i])
    x.append(cnt)
    cnt += 1
print(len(data))

plt.figure(figsize=(10, 8))
plt.style.use('default')
plt.plot(x,y_true,color='blue', label='Actual Training Run')

plt.plot(x,y_predit,color='red', label='Simulated Training Run')
plt.xlabel("Training Steps")
plt.ylabel("Loss of Test Example")
plt.legend()

plt.grid(True)
plt.show()
```
